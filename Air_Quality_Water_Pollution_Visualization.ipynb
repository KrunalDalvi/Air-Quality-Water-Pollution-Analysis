```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'cities_air_quality_water_pollution.csv')
df.head()
```

```python
#Size of Data
df.shape
```

```python
#Dataset Overview and Summary
df.info()
```

```python
# Descriptive Statistics
df.describe()
```

```python
# List of Column Names
df.columns
```

```python
# Clean and Format Column Names
df.columns = df.columns.str.replace('"','').str.title()
df.columns
```

```python
# Count of Missing Values
df.isnull().sum() 
```

```python
# Identify Duplicate Rows
df.duplicated()
```

```python
# Top 10 countries with clean water
clean_water_top10 = df.groupby('  Country ')['  Waterpollution '].mean().nsmallest(10).reset_index()

# Plotting
plt.figure(figsize=(14, 10))

# Bar graph for Top 10 countries with clean water
plt.subplot(2, 2, 1)
sns.barplot(x='  Waterpollution ', y='  Country ', data=clean_water_top10, palette='viridis')
plt.title('Top 10 Countries with Clean Water')
plt.xlabel('Water Pollution Index')

plt.tight_layout()
plt.show()
```

Croatia and Singapore are noted for their clean water due to stringent environmental regulations, efficient water management systems, ensuring high-quality water sources that meet international standards and support public health and sanitation.

```python
# Top 10 countries with dirty water
dirty_water_top10 = df.groupby('  Country ')['  Waterpollution '].mean().nlargest(10).reset_index()

# Plotting
plt.figure(figsize=(14, 10))

# Bar graph for Top 10 countries with dirty water
plt.subplot(2, 2, 2)
sns.barplot(x='  Waterpollution ', y='  Country ', data=dirty_water_top10, palette='magma')
plt.title('Top 10 Countries with Dirty Water')
plt.xlabel('Water Pollution Index')

plt.tight_layout()
plt.show()
```

The prevalence of dirty water in Benin and the Central African Republic is primarily due to inadequate infrastructure for clean water supply, poor sanitation practices, limited access to safe drinking water sources, and environmental degradation, leading to health hazards and waterborne diseases among their populations.

```python
# Top 10 countries with good air quality
good_air_top10 = df.groupby('  Country ')['  Airquality '].mean().nlargest(10).reset_index()

# Plotting
plt.figure(figsize=(14, 10))

# Bar graph for Top 10 countries with good air quality
plt.subplot(2, 2, 3)
sns.barplot(x='  Airquality ', y='  Country ', data=good_air_top10, palette='viridis')
plt.title('Top 10 Countries with Good Air Quality')
plt.xlabel('Air Quality Index')

plt.tight_layout()
plt.show()
```

Countries like Eritrea and Federated States of Micronesia Republic may have good air quality due to low industrialization, limited vehicle emissions, sparse population density, and minimal urban development, leading to reduced pollution levels and cleaner atmospheric conditions.

```python
# Top 10 countries with bad air quality
bad_air_top10 = df.groupby('  Country ')['  Airquality '].mean().nsmallest(10).reset_index()

# Plotting
plt.figure(figsize=(14, 10))

# Bar graph for Top 10 countries with bad air quality
plt.subplot(2, 2, 4)
sns.barplot(x='  Airquality ', y='  Country ', data=bad_air_top10, palette='magma')
plt.title('Top 10 Countries with Bad Air Quality')
plt.xlabel('Air Quality Index')

plt.tight_layout()
plt.show()
```

Countries like Angola and Kazakhstan often experience poor air quality due to factors such as industrial emissions, reliance on fossil fuels for energy production, inadequate environmental regulations, and geographical conditions that trap pollutants, exacerbating air pollution levels in these regions.

```python
plt.figure(figsize=(14, 6))

# Histogram for Water Pollution
plt.subplot(1, 2, 2)
sns.histplot(df['  Waterpollution '], kde=True, color='red', bins=30)
plt.title('Distribution of Water Pollution')
plt.xlabel('Water Pollution Index')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Histogram for Air Quality
plt.subplot(1, 2, 1)
sns.histplot(df['  Airquality '], kde=True, color='blue', bins=30)
plt.title('Distribution of Air Quality')
plt.xlabel('Air Quality Index')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
